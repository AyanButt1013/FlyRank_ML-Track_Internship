## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

### **What I CAN Claim**
* **Observed Correlations:** I can identify statistical associations between observable content attributes (e.g., word count, content recency) and search performance measurements (e.g., impressions, CTR, position tiers).
* **Directional Insights:** I can highlight broader patterns—such as whether stale content (180+ days old) or thin pages tend to exhibit lower visibility across the dataset.
* **Decision Support & Prioritization:** I can deliver an evidence-backed ranking framework to help editorial and SEO teams prioritize which candidate pages to review first based on empirical risk signals.

### **What I CANNOT Claim**
* **Causal Proof:** I cannot claim that updating or expanding a page will directly *cause* a search ranking recovery, as observational metrics without controlled experiments cannot establish causality.
* **"Predicting Google" or Algorithm Rules:** I cannot claim to have reverse-engineered Google's algorithm, underlying ranking factors, or AI referral systems.
* **Universal Truths:** I cannot claim that these observed relationships hold true outside the specific pseudonymized snapshot analyzed in this project.
